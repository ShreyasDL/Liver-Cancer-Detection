```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pydicom as dicom
%matplotlib inline
```

```python
def dicom_read(path) :
    ds=dicom.dcmread(path)
    dcm_sample=ds.pixel_array*128
    return dcm_sample
```

```python
def display(img,cmap='gray') :
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('JPEG/image_63 CT p11.jpg')
```

```python
display(img)
```

```python
pat = cv2.imread('JPEG\image_63 liver p11.jpg')
```

```python
display(pat)
```

```python
orb = cv2.ORB_create()
```

```python
kp1,des1 = orb.detectAndCompute(img,None)
kp2,des2 = orb.detectAndCompute(pat,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck = True)
```

```python
matches = bf.match(des1,des2)
```

```python
matches = sorted(matches,key=lambda x:x.distance)
```

```python
liver_match = cv2.drawMatches(img,kp1,pat,kp2,matches[:25],None,flags=2)
```

```python
display(liver_match)
```
