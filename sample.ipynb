```python
import numpy as np
import cv2
import pydicom as dicom
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
ds=dicom.dcmread('DCM\image_63 CT p11.dcm')
ct_liver = ds.pixel_array
# cv2.imwrite('new.jpg',ct_liver)
# a = cv2.imread('new.jpg')
# print(a.shape)
a = np.zeros((512,512,3),np.int16)
a[:,:,0] = ct_liver
a[:,:,1] = ct_liver
a[:,:,2] = ct_liver
a = np.uint8(a)
a.dtype
```

```python
# APPLY BLUR
img = cv2.medianBlur(a,3) # DISPLAY
```

```python
display(img)
```
