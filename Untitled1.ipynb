```python
import pydicom as dicom
import numpy as np
ds=dicom.dcmread("DCM\image_63 CT p11.dcm")
# store the raw image data
img = ds.pixel_array

rescale_slope=1 
rescale_intercept=-1024

def translate(value,rescale_slope,rescale_intercept):

    return (value*rescale_slope)+rescale_intercept 

def int12_to_int8(DicomImage):
    img_array = []

    for eachRow in DicomImage:
        for eachPix in eachRow:
            img_array.append(translate(eachPix,rescale_slope,rescale_intercept))
    img_array = np.array(img_array)
    img_array = img_array.reshape(512,512)  
    return img_array

img_1 = int12_to_int8(img)
```

```python
import numpy as np
import cv2
import pydicom as dicom
import matplotlib.pyplot as plt
%matplotlib inline

def display(img,cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
def get_LUT_value(data, window, level):

    return np.piecewise(data, 
        [data <= (level - 0.5 - (window-1)/2),
            data > (level - 0.5 + (window-1)/2)],
            [0, 255, lambda data: ((data - (level - 0.5))/(window-1) + 0.5)*(255-0)])

level=200
window=800

scaled_img=get_LUT_value(img, window, level)
```

```python
scaled_img = cv2.convertScaleAbs(scaled_img)
```

```python
img = np.zeros((512,512,3),np.uint8)
img[:,:,0] = scaled_img
img[:,:,1] = scaled_img
img[:,:,2] = scaled_img
img.shape
display(img)
```

```python
# APPLY BLUR
img = cv2.medianBlur(img,3) # DISPLAY
```

```python
# CONVERT TO GRAY SCALE
# gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
a = np.zeros((512,512),np.uint8)
a = img
a.shape
```

```python
# APPLY THRESHOLD NOT BINARY ==> APPLY OTSU'S METHOD
ret, thresh = cv2.threshold(a,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)  # DISPLAY
```

```python
display(thresh)
```

```python
# NOISE REMOVAL (OPTIONAL)
kernel = np.ones((3,3),np.uint8)
```

```python
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel,iterations = 2)     # DISPLAY
```

```python
display(opening)
```

```python
sure_bg = cv2.dilate(opening,kernel,iterations=3)  # DISPLAY
```

```python
# DISTANCE TRANSFORM
dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,5) # DISPLAY
```

```python
ret, sure_fg = cv2.threshold(dist_transform,0.7*dist_transform.max(),255,0) # DISPLAY
display(dist_transform)
```

```python
sure_fg = np.uint8(sure_fg)
```

```python
unknown = cv2.subtract(sure_bg,sure_fg) # DISPLAY
```

```python
# CREATING MARKERS => 3 STEPS

# 1. GETTING MARKERS

ret, markers = cv2.connectedComponents(sure_fg)
```

```python
markers = markers + 1
```

```python
markers[unknown == 255] = 0  # DISPLAY
```

```python
markers = cv2.watershed(a,markers)   # DISPLAY
```
